```python
import pandas as pd

print(pd.__version__)
```

```python
df = pd.read_csv(csv_url)
```

```python
df.head()
```

Country

```python
df["Country"] = df["Location"].str.extract(r"^(.*?)\s*\(", expand=False)
```

```python
print(df[["Location", "Country"]].head())
```

Date

```python
df["Update_Date"] = df["Location"].str.extract(
    r"\(as of (.*?)\)",
    expand=False
)
```

```python
print(df[["Location", "Update_Date"]].head())
```

```python
df["Update_Date"] = pd.to_datetime(df["Update_Date"])
```

```python
print(df[["Country", "Update_Date"]].head())
```

```python
country_df = df[["Country", "Update_Date", "Reported", "Count"]]
```

```python
country_df
```

Remove totals

```python
country_df = country_df.dropna(subset=["Country"])
```

```python
country_df
```

Storing in data base

```python
import sqlite3

connection = sqlite3.connect("ebola_surveillance.db")
cursor = connection.cursor()
```

countries table

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS countries (

    country_id INTEGER PRIMARY KEY AUTOINCREMENT,

    country_name TEXT NOT NULL UNIQUE

)
""")
```

report type table

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS report_types (

    report_type_id INTEGER PRIMARY KEY AUTOINCREMENT,

    report_name TEXT NOT NULL UNIQUE

)
""")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS ebola_reports (

    report_id INTEGER PRIMARY KEY AUTOINCREMENT,

    country_id INTEGER NOT NULL,

    report_type_id INTEGER NOT NULL,

    update_date DATE NOT NULL,

    count INTEGER NOT NULL,

    FOREIGN KEY (country_id)
        REFERENCES countries(country_id),

    FOREIGN KEY (report_type_id)
        REFERENCES report_types(report_type_id)
)
""")
```

```python
connection.commit()
```

```python
for index, row in country_df.iterrows():

    country = row["Country"]
    report = row["Reported"]
    update_date = row["Update_Date"]
    count = row["Count"]

    cursor.execute("""
    SELECT country_id
    FROM countries
    WHERE country_name = ?
    """, (country,))

    country_id = cursor.fetchone()[0]

    cursor.execute("""
    SELECT report_type_id
    FROM report_types
    WHERE report_name = ?
    """, (report,))

    report_type_id = cursor.fetchone()[0]

    update_date = update_date.strftime("%Y-%m-%d")

    cursor.execute("""
    INSERT INTO ebola_reports
    (country_id, report_type_id, update_date, count)

    VALUES (?, ?, ?, ?)
    """, (
        country_id,
        report_type_id,
        update_date,
        count
    ))

connection.commit()
```

```python
cursor.execute("""
SELECT
    c.country_name,
    r.report_name,
    e.update_date,
    e.count
FROM ebola_reports e
JOIN countries c
    ON e.country_id = c.country_id
JOIN report_types r
    ON e.report_type_id = r.report_type_id
ORDER BY c.country_name, r.report_type_id;
""")

for row in cursor.fetchall():
    print(row)
```

```python
cursor.execute("""
SELECT
    c.country_name,
    e.count

FROM ebola_reports e

JOIN countries c
ON e.country_id = c.country_id

JOIN report_types r
ON e.report_type_id = r.report_type_id

WHERE r.report_name = 'Confirmed cases'

ORDER BY e.count DESC;
""")

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute("""
SELECT
    c.country_name,
    e.count

FROM ebola_reports e

JOIN countries c
ON e.country_id = c.country_id

JOIN report_types r
ON e.report_type_id = r.report_type_id

WHERE r.report_name = 'Probable cases'

ORDER BY e.count DESC;
""")

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute("""
SELECT
    c.country_name,
    COUNT(*) AS number_of_reports

FROM ebola_reports e

JOIN countries c
ON e.country_id = c.country_id

JOIN report_types r
ON e.report_type_id = r.report_type_id

WHERE r.report_name = 'Confirmed cases'

GROUP BY c.country_name;
""")

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
import sqlite3

connection = sqlite3.connect("ebola_surveillance.db")
cursor = connection.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS symptom_reports (

    report_id INTEGER PRIMARY KEY AUTOINCREMENT,

    person_name TEXT,

    country TEXT,

    symptoms TEXT,

    symptom_start_date TEXT,

    travel_history TEXT,

    contact_history TEXT,

    date_submitted TIMESTAMP DEFAULT CURRENT_TIMESTAMP

)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS exposure_reports (

    exposure_id INTEGER PRIMARY KEY AUTOINCREMENT,

    person_name TEXT,

    country TEXT,

    exposure_type TEXT,

    exposure_date TEXT,

    comments TEXT,

    date_submitted TIMESTAMP DEFAULT CURRENT_TIMESTAMP

)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS public_health_messages (

    message_id INTEGER PRIMARY KEY AUTOINCREMENT,

    name TEXT,

    email TEXT,

    country TEXT,

    subject TEXT,

    message TEXT,

    date_submitted TIMESTAMP DEFAULT CURRENT_TIMESTAMP

)
""")

connection.commit()
connection.close()

print("Tables created successfully!")
```
